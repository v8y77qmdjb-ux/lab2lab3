```python
import pandas as pd

data = [
    ["apple", "táo"],
    ["book", "sách"],
    ["water", "nước"],
    ["computer", "máy tính"],
    ["school", "trường học"]
]

df = pd.DataFrame(data, columns=["word", "meaning"])
df.to_csv("vocab.csv", index=False)

print("Đã tạo file vocab.csv")
```

```python
import pandas as pd
import random

df = pd.read_csv("vocab.csv")
words = df.values.tolist()

random.shuffle(words)

score = 0

for word, meaning in words:
    print(f"Từ: {word}")
    answer = input("Nghĩa là gì? ")

    if answer.lower() == "exit":
        break

    if answer.strip().lower() == meaning.lower():
        print("Đúng!\n")
        score += 1
    else:
        print(f"Sai! Đáp án: {meaning}\n")

print(f"Điểm: {score}")
```
